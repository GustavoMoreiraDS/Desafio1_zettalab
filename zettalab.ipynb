# Dados Desmatamento 游꺕


## Importa칞칚o e tratamento inicial dos dados

```python
dados_desmat = pd.read_csv('/content/bquxjob_4d75a40d_196a6d851ce.csv') #importa칞칚o dos dados do tipo csv para o ambiente
dados_desmat.head() #visuliza칞칚o incial da tabela com os primeiros 5 valores com a fun칞칚o head
```

```python
dados_desmat['id_municipio'] = dados_desmat['id_municipio'].astype(str) #transformar a coluna para str
dados_desmat = dados_desmat.loc[dados_desmat['id_municipio'].str.startswith('15')] # separar apenas os que come칞am com 15 pois representam os municipios no estado do par치
dados_desmat.head() #vizualia칞칚o inicial
```

## Maior porcentagem desmatada em 2023

```python
dados_desmat['ano'].unique() # visualiza칞칚o dos valores 칰nicos, nos leva a concluir que o 칰ltimo ano dos dados 칠 2023

dados_desmat2023 = dados_desmat[dados_desmat['ano'] == 2023] # filtro para pegar apenas os dados do ano de 2023

dados_desmat2023 = dados_desmat2023[dados_desmat2023['desmatado'] != 0] # removendo linhas em que a 치rea desmatada 칠 zero, o que causaria divis칚o por zero e porcentagem infinita (aconteceu anteriormente e adicionei essa linha)

dados_desmat2023['porcentagem_desmatada'] = (dados_desmat2023['desmatado'] / dados_desmat2023['area_total']) * 100  # criando a nova coluna com a porcentagem entre 치rea total e 치rea desmatada no ano de 2023

dados_desmat2023.head() #verificando se funcionou

top10 = dados_desmat2023.sort_values('porcentagem_desmatada', ascending= False).head(10) # definindo os top 10 municipios com porcentagem desmatada
top10 #visualizando os 10
```

## Maior 치rea desmatada

```python
'''visualiza칞칚o dos top 15 municipios com maior 치rea desmatada usando a fun칞칚o sort_values, escolhendo a
coluna que depresenta a 치rea desmatada e ordenando do maior para o menor'''
dados_desmat2023.sort_values('desmatado', ascending= False).head(15)
```

## Municipios com maior crescimento do desmatamento em porcentagem

```python
# filtrando apenas os dados dos anos de 2000 e 2023
dados_2000 = dados_desmat[dados_desmat['ano'] == 2000][['id_municipio_nome', 'desmatado']]
dados_2023 = dados_desmat[dados_desmat['ano'] == 2023][['id_municipio_nome', 'desmatado']]


dados_comparativo = pd.merge(dados_2000, dados_2023, on='id_municipio_nome', suffixes=('_2000', '_2023')) # juntando os dados dos dois anos em um 칰nico DataFrame


dados_comparativo = dados_comparativo[dados_comparativo['desmatado_2000'] != 0] # removendo casos onde o valor de desmatamento em 2000 칠 zero (evita divis칚o por zero)


dados_comparativo['variacao_percentual'] = ((dados_comparativo['desmatado_2023'] - dados_comparativo['desmatado_2000']) / dados_comparativo['desmatado_2000']) * 100
# calculando a varia칞칚o percentual do desmatamento de 2000 para 2023


dados_comparativo.sort_values('variacao_percentual', ascending=False).head(10) # exibindo os 10 munic칤pios com maior aumento percentual de desmatamento
```

## Anos com maior aumento de desmatate por 치rea.

```python
dados_desmat.groupby('ano')['desmatado'].sum() # somando a 치rea desmatada por ano, mas n칚o faz sentido uma vez que se avalio o ano de 2001 ele j치 conta do o desmate do ano 2000
desmate_por_ano = dados_desmat.groupby('ano')['desmatado'].sum().sort_index() # atribuindo a um objeto
variacao_anual = desmate_por_ano.diff()  # calculando a diferen칞a ano a anp
variacao_anual # visualizando se deu certo


tabela_variacao = pd.DataFrame({                         # comparar os dataframes
    'desmatado_total': desmate_por_ano,
    'variacao_em_relacao_ao_ano_anterior': variacao_anual
})

tabela_variacao.to_csv('variacao_desmatamento_por_ano.csv', index=True) #exportando a tabela

tabela_variacao # visualiza칞칚o completa da tabela

tabela_variacao.sort_values('variacao_em_relacao_ao_ano_anterior', ascending=False) # ordenando para os anos com o maior desmate
# podemos concluir que entre os anos de 2001 e 2005 tivemos uma maior expans칚o da 치rea desmatada. em 2000 n칚o temos como calcular por falta de dados em 1999
```

#Dados Homic칤dio 游뽖
