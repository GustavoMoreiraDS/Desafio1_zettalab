# Dados Desmatamento 🌳


## Importação e tratamento inicial dos dados

```python
dados_desmat = pd.read_csv('/content/bquxjob_4d75a40d_196a6d851ce.csv') #importação dos dados do tipo csv para o ambiente
dados_desmat.head() #visulização incial da tabela com os primeiros 5 valores com a função head
```

```python
dados_desmat['id_municipio'] = dados_desmat['id_municipio'].astype(str) #transformar a coluna para str
dados_desmat = dados_desmat.loc[dados_desmat['id_municipio'].str.startswith('15')] # separar apenas os que começam com 15 pois representam os municipios no estado do pará
dados_desmat.head() #vizualiação inicial
```

## Maior porcentagem desmatada em 2023

```python
dados_desmat['ano'].unique() # visualização dos valores únicos, nos leva a concluir que o último ano dos dados é 2023

dados_desmat2023 = dados_desmat[dados_desmat['ano'] == 2023] # filtro para pegar apenas os dados do ano de 2023

dados_desmat2023 = dados_desmat2023[dados_desmat2023['desmatado'] != 0] # removendo linhas em que a área desmatada é zero, o que causaria divisão por zero e porcentagem infinita (aconteceu anteriormente e adicionei essa linha)

dados_desmat2023['porcentagem_desmatada'] = (dados_desmat2023['desmatado'] / dados_desmat2023['area_total']) * 100  # criando a nova coluna com a porcentagem entre área total e área desmatada no ano de 2023

dados_desmat2023.head() #verificando se funcionou

top10 = dados_desmat2023.sort_values('porcentagem_desmatada', ascending= False).head(10) # definindo os top 10 municipios com porcentagem desmatada
top10 #visualizando os 10
```

## Maior área desmatada

```python
'''visualização dos top 15 municipios com maior área desmatada usando a função sort_values, escolhendo a
coluna que depresenta a área desmatada e ordenando do maior para o menor'''
dados_desmat2023.sort_values('desmatado', ascending= False).head(15)
```

## Municipios com maior crescimento do desmatamento em porcentagem

```python
# filtrando apenas os dados dos anos de 2000 e 2023
dados_2000 = dados_desmat[dados_desmat['ano'] == 2000][['id_municipio_nome', 'desmatado']]
dados_2023 = dados_desmat[dados_desmat['ano'] == 2023][['id_municipio_nome', 'desmatado']]


dados_comparativo = pd.merge(dados_2000, dados_2023, on='id_municipio_nome', suffixes=('_2000', '_2023')) # juntando os dados dos dois anos em um único DataFrame


dados_comparativo = dados_comparativo[dados_comparativo['desmatado_2000'] != 0] # removendo casos onde o valor de desmatamento em 2000 é zero (evita divisão por zero)


dados_comparativo['variacao_percentual'] = ((dados_comparativo['desmatado_2023'] - dados_comparativo['desmatado_2000']) / dados_comparativo['desmatado_2000']) * 100
# calculando a variação percentual do desmatamento de 2000 para 2023


dados_comparativo.sort_values('variacao_percentual', ascending=False).head(10) # exibindo os 10 municípios com maior aumento percentual de desmatamento
```

## Anos com maior aumento de desmatate por área.

```python
dados_desmat.groupby('ano')['desmatado'].sum() # somando a área desmatada por ano, mas não faz sentido uma vez que se avalio o ano de 2001 ele já conta do o desmate do ano 2000
desmate_por_ano = dados_desmat.groupby('ano')['desmatado'].sum().sort_index() # atribuindo a um objeto
variacao_anual = desmate_por_ano.diff()  # calculando a diferença ano a anp
variacao_anual # visualizando se deu certo


tabela_variacao = pd.DataFrame({                         # comparar os dataframes
    'desmatado_total': desmate_por_ano,
    'variacao_em_relacao_ao_ano_anterior': variacao_anual
})

tabela_variacao.to_csv('variacao_desmatamento_por_ano.csv', index=True) #exportando a tabela

tabela_variacao # visualização completa da tabela

tabela_variacao.sort_values('variacao_em_relacao_ao_ano_anterior', ascending=False) # ordenando para os anos com o maior desmate
# podemos concluir que entre os anos de 2001 e 2005 tivemos uma maior expansão da área desmatada. em 2000 não temos como calcular por falta de dados em 1999
```

#Dados Homicídio 🩸

## Importação dos dados

```python
dados_homicidio = pd.read_excel('/content/ipeadata[06-05-2025-02-57].xls') #importação do arquivo xls
dados_homicidio.head() # visualização inicial e verificar se a importação foi correta. com ela podemos verificar muitos valores faltantes na decada de 80
```

## Anos mais violentos

```python
colunas_anos = dados_homicidio.columns[3:] #selecionando apenas as colunas dos anos
dados_homicidio[colunas_anos].sum().sort_values(ascending= False).head()
#acessando dentro do df apenas as colunas de anos e somando seus valores e ordenando do maior para o menor. usando head para saber apenas os 5 primeiros'''

#concluindo que os anos 2017,2018,2016 foram os tres mais violentos do estado. valores coerentes com notícias que registram que foram os anos mais violentos do estado
#seguidos de uma redução expressiva no número de homicidios por conta de reforços no policiamento
```

## Ano mais violento de cada município

```python
# usando a colunas_anos já criada, usamos idmax para encontrar o maior valor entre os anos e adicionar a nova coluna do df chamada ano_mais_violento'''
dados_homicidio['ano_mais_violento'] = dados_homicidio[colunas_anos].idxmax(axis=1)

dados_homicidio[['Município', 'ano_mais_violento']] # visualização do municipio e do ano mais violento, poderia colocar o df inteiro mas teria muito informação

dados_homicidio.to_excel('dados_homicidio_com_ano_mais_violento.xlsx', index=False)

# consulta do ano mais violento ↓             Municipio a escolha↓
dados_homicidio[dados_homicidio['Município'] == 'São Félix do Xingu'][['Município', 'ano_mais_violento']]
#em são felix do xingu, um dos estados com maior área desmatada temos o ano de 2016 como o mais violento
```
